```python
import pandas as pd
```

```python
df = pd.read_csv("Resources/COVID19VaccinesRecords.csv", low_memory=False)
```

```python
df.head()
```

```python
# Delete the first row
df = df.iloc[1:]
```

```python
df.head()
```

```python
df = df.drop(columns=["COVID-19Vaccines"])
```

```python
df.head()
```

```python
df.rename(columns={"as_of_date": "As Of Date"}, inplace=True)
df.rename(columns={"zip_code_tabulation_area": "Zip Code"}, inplace=True)
df.rename(columns={"local_health_jurisdiction": "Local Health Jurisdiction"}, inplace=True)
df.rename(columns={"county": "County"}, inplace=True)
df.rename(columns={"vaccine_equity_metric_quartile": "Vaccine Equity Metric Quartile"}, inplace=True)
df.rename(columns={"vem_source": "Vem Source"}, inplace=True)
df.rename(columns={"age12_plus_population": "Age 12+ Population"}, inplace=True)
df.rename(columns={"age5_plus_population": "Age 5+ Population"}, inplace=True)
df.rename(columns={"tot_population": "Total Population"}, inplace=True)
df.rename(columns={"persons_fully_vaccinated": "People Fully Vaccinated"}, inplace=True)
df.rename(columns={"persons_partially_vaccinated": "People Partially Vaccinated"}, inplace=True)
df.rename(columns={"percent_of_population_fully_vaccinated": "Percent Of Population Fully Vaccinated"}, inplace=True)
df.rename(columns={"percent_of_population_partially_vaccinated": "Percent Of Population Partially Vaccinated"}, inplace=True)
df.rename(columns={"percent_of_population_with_1_plus_dose": "Percent of Population With 1+ Dose"}, inplace=True)
df.rename(columns={"up_to_date_count": "Up To Date Count"}, inplace=True)
```
